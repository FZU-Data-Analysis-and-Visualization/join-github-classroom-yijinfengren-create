### 六、合并与拼接

1. 横向合并：

   df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})

   df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})

   按 id 合并。

```python
df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})

pd.merge(df1, df2, on="id")
```

2. 纵向拼接：

   df_a = pd.DataFrame({"val": [1, 2]})

   df_b = pd.DataFrame({"val": [3, 4]})

   拼接后重置索引。

```python
df_a = pd.DataFrame({"val": [1, 2]})
df_b = pd.DataFrame({"val": [3, 4]})

pd.concat([df_a, df_b], ignore_index=True)
```

### 七、时间处理

1. 创建时间列：["2025-01-01", "2025-01-02", "2025-01-03"]，转为 datetime 格式。
2. 提取时间列的月份、星期。

```python
df_time = pd.DataFrame({"date": ["2025-01-01", "2025-01-02", "2025-01-03"]})
df_time["date"] = pd.to_datetime(df_time["date"])
df_time
```

```python
weekday_map = {
    0: "星期一",
    1: "星期二",
    2: "星期三",
    3: "星期四",
    4: "星期五",
    5: "星期六",
    6: "星期日"
}

df_time["month"] = df_time["date"].dt.month
df_time["weekday"] = df_time["date"].dt.dayofweek.map(weekday_map)
df_time
```

### 八、高级操作

1. 对 score 列做标准化：`(值 - 均值) / 标准差`。
2. 用透视表：行是 gender，值是 score，聚合函数为中位数。

```python
df["score_standardized"] = (df["score"] - df["score"].mean()) / df["score"].std()
df
```

```python
pd.pivot_table(df, index="gender", values="score", aggfunc="median")
```
